```python
%matplotlib inline

import numpy
from matplotlib import pyplot
import matplotlib.ticker

import subprocess
import os.path
```

```python
data = subprocess.Popen(['find', os.path.expanduser("~"), '-printf', '%s\n'], stdout=subprocess.PIPE)

arr = numpy.genfromtxt(data.stdout, dtype=int)

arr.sort()
```

```python
fig, ax = pyplot.subplots()
fig.set_size_inches(10, 6)
fig.set_dpi(100)
ax.semilogy(arr)
ax.set_xticks([])
ax.yaxis.set_major_formatter(matplotlib.ticker.FixedFormatter(["1KB", "1MB", "1GB"]))
ax.set_yticks([1000, 1000000, 1000000000])
ax.yaxis.set_label_text("Size in bytes")
ax.grid()
pyplot.savefig("/tmp/output.png")
```
